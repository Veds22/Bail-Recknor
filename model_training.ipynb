```python
import pandas as pd
import numpy as np
from sklearn.metrics import recall_score, classification_report

```

```python
X_train = pd.read_csv('./dataset/X_train_final.csv')
X_test = pd.read_csv('./dataset/X_test_final.csv')
y_train = pd.read_csv('./dataset/y_train.csv')
y_test = pd.read_csv('./dataset/y_test.csv')
```

```python
X_train.head()
```

Criteria of Training: Recall

Models tested:
    Logrithmic Regression
    Linear SVM
    Simple Feed-Forward Neural Network



```python
# MODEL 1: Logrithmic regression
from sklearn.linear_model import LogisticRegression

log_reg =  LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)
log_reg.fit(X_train, y_train)
```

```python
X_test.head(50)
```

```python
y_pred_lr = log_reg.predict(X_test)

print("Logistic Regression Recall:",
      recall_score(y_test, y_pred_lr))

print(classification_report(y_test, y_pred_lr))
```

```python
# MODEL 2: Linear SVM
from sklearn.svm import LinearSVC

lin_svm = LinearSVC(
    class_weight='balanced',
    random_state=42
)
    
lin_svm.fit(X_train, y_train)
y_pred_svm = lin_svm.predict(X_test)

print("Linear SVM Recall:",
      recall_score(y_test, y_pred_svm))

print(classification_report(y_test, y_pred_svm))
```

```python
# MODEL 3: MLP Classifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
from sklearn.metrics import classification_report

# Convert sparse matrix to dense (MLP needs dense input)
X_train_nn = X_train.values
X_test_nn = X_test.values

# Model
mlp = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_nn.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

mlp.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Early stopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train
mlp.fit(
    X_train_nn,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Predict
y_pred_mlp = (mlp.predict(X_test_nn) > 0.5).astype(int)

# Evaluation
print("MLP Results")
print(classification_report(y_test, y_pred_mlp))
```

```python
from sklearn.metrics import recall_score

y_pred_mlp = (mlp.predict(X_test_nn) > 0.5).astype(int)

print("MLP Recall:",
      recall_score(y_test, y_pred_mlp))

print(classification_report(y_test, y_pred_mlp))

```
